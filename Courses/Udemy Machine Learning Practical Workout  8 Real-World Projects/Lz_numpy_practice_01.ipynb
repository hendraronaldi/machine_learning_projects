### XOR Exercise

```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
N = 5000
x = np.random.random((N, 2)) * 2 - 1
y = np.zeros(N)
y[(x[:, 0] < 0) & (x[:, 1] > 0)] = 1
y[(x[:, 0] > 0) & (x[:, 1] < 0)] = 1
```

```
plt.scatter(x[:, 0], x[:, 1], c=y)
```

### Donut exercise

```
import pandas as pd
```

```
df = pd.DataFrame({
    'x1': x[:,0],
    'x2': x[:,1],
    'y': y
})
```

```
def quadratic_rows(row):
    return row**2
```

```
def mul_col(row):
    return row['x1'] * row['x2']
```

```
df['x1^2'] = df['x1'].apply(quadratic_rows)
df['x2^2'] = df['x2'].apply(quadratic_rows)
df['x1*x2'] = df.apply(mul_col, axis=1)
df.head()
```

### Convo exercise

```
from PIL import Image
```

```
im = Image.open('/content/5b8e043c31e1380004315b5f.png')
```

```
arr = np.array(im)
```

```
gray = arr.mean(axis=2)
```

```
from scipy.signal import convolve2d
```

```
hx = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
hy = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])

gx = convolve2d(hx, gray)
gy = convolve2d(hy, gray)
```

```
g = np.sqrt(gx**2 + gy**2)
```

```
plt.imshow(g)
```
