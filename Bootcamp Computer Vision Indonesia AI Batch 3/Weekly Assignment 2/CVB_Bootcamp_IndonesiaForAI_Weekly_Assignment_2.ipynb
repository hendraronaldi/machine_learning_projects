```python
Missing_value = pd.DataFrame({"Count":df.isnull().sum(),"Percentage":round(100*df.isnull().sum()/df.count().max(), 2)})

Missing_value.T
```

To handle missing data, we remove rows from the dataset that contain more than 70% NaN values. This step is crucial to avoid potential errors in the final results.

```python
df=df.dropna(thresh=0.7*len(df.columns))
```

```python
df.shape
```

We replace the remaining missing data values in each column with the average value of the data (mean) for that specific column. Additionally, to enhance accuracy, we compute the average of the output data based on the class labels (Outcome).

```python
for i in df.columns[df.isnull().any(axis=0)]:
    df[i] = df.groupby(["Outcome"])[i].transform(lambda x: x.fillna(x.mean()))
```

## Outlier Handling

```python
Q1  = pd.DataFrame()
Q3  =   pd.DataFrame()
IQR =  pd.DataFrame()
UpperLimit =  pd.DataFrame()
LowerLimit =  pd.DataFrame()
Outlier_UpperLimit =  pd.DataFrame(index=[1,2,3,4])
Outlier_LowerLimit =  pd.DataFrame(index=[1,2,3,4])

columns = df.columns[0:-1]

for i in columns:
    Q1[i] = [df[i].quantile(0.25)]
    Q3[i]=  [df[i].quantile(0.75)]
    IQR[i] = [Q3[i][0]-Q1[i][0]]
    UpperLimit[i] = [Q3[i][0]+1.5*IQR[i][0]]
    LowerLimit[i] = [Q1[i][0]-1.5*IQR[i][0]]
```

```python
dfs = pd.concat([Q1,Q3,IQR,UpperLimit,LowerLimit])
dfs.index = ["Q1","Q3","IQR","UpperLimit","LowerLimit"]
dft = dfs.T
dft
```

Since the target is imbalanced class, let's try use f1 score, precision, and recall for metrics since accuracy can be misleading

```python
import tensorflow as tf
from bayes_opt import BayesianOptimization
from sklearn.metrics import f1_score
from keras.metrics import Precision, Recall, Accuracy
```

```python
# Define a custom F1 score metric function that works with TensorFlow tensors
def f1_metric(y_true, y_pred):
    y_true = tf.convert_to_tensor(y_true, dtype=tf.float32)
    y_pred = tf.convert_to_tensor(y_pred, dtype=tf.float32)

    true_positives = tf.reduce_sum(tf.math.round(tf.clip_by_value(y_true * y_pred, 0, 1)))
    possible_positives = tf.reduce_sum(tf.math.round(tf.clip_by_value(y_true, 0, 1)))
    predicted_positives = tf.reduce_sum(tf.math.round(tf.clip_by_value(y_pred, 0, 1)))

    precision = true_positives / (predicted_positives + tf.keras.backend.epsilon())
    recall = true_positives / (possible_positives + tf.keras.backend.epsilon())
    f1 = 2 * (precision * recall) / (precision + recall + tf.keras.backend.epsilon())

    return f1
```

```python
# Define the neural network architecture
def create_model(learning_rate, num_hidden_layers, num_neurons):
    model = tf.keras.Sequential()
    for i in range(num_hidden_layers):
        if i == 0:
            model.add(tf.keras.layers.Dense(num_neurons, input_dim=X_train.shape[1], activation='relu'))
        else:
            model.add(tf.keras.layers.Dense(num_neurons, activation='relu'))
    model.add(tf.keras.layers.Dense(1, activation='sigmoid'))
    optimizer = tf.keras.optimizers.Adam(learning_rate=learning_rate)
    model.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=[f1_metric, Precision(), Recall(), 'accuracy'])
    return model
```

```python
# Define the objective function to optimize
def objective(learning_rate, epochs, num_hidden_layers, num_neurons):
    model = create_model(learning_rate, int(num_hidden_layers), int(num_neurons))
    model.fit(X_train, y_train, epochs=int(epochs), validation_data=(X_test, y_test), verbose=0)
    val_loss, val_f1, val_precision, val_recall, val_accuracy = model.evaluate(X_test, y_test, verbose=0)
    # history = model.fit(X_train, y_train, epochs=int(epochs), validation_split=0.1, verbose=0)
    # val_f1 = history.history['val_f1_metric'][-1]
    return val_recall
```

```python
# Define the search space for hyperparameters
pbounds = {'learning_rate': (0.0001, 0.1),
           'num_hidden_layers': (1, 5),
           'num_neurons': (5, 50),
           'epochs': (5, 50)}
```

```python
# Perform Bayesian optimization
optimizer = BayesianOptimization(f=objective, pbounds=pbounds, verbose=2)
optimizer.maximize(init_points=5, n_iter=20)
```

```python
# Print the optimized hyperparameters and validation f1 score
print('Optimized hyperparameters:')
print(optimizer.max['params'])
print('Validation f1 score: {:.2f}%'.format(optimizer.max['target'] * 100))
```

```python
def plot_confusion_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    class_labels = ["0", "1"]

    # Create a figure and axis for the plot
    fig, ax = plt.subplots()

    # Create a heatmap of the confusion matrix
    im = ax.imshow(cm)

    # Display color bar
    cbar = ax.figure.colorbar(im, ax=ax)

    # Show all ticks and label them with the respective list entries
    ax.set_xticks(np.arange(len(class_labels)))
    ax.set_yticks(np.arange(len(class_labels)))
    # Label the ticks with the respective class labels
    ax.set_xticklabels(class_labels)
    ax.set_yticklabels(class_labels)

    # Loop over data dimensions and create text annotations
    for i in range(len(class_labels)):
        for j in range(len(class_labels)):
            text = ax.text(j, i, cm[i, j], ha="center", va="center", color="r")

    # Set axis labels
    ax.set_xlabel("Predicted Labels")
    ax.set_ylabel("True Labels")

    # Set title
    ax.set_title("Confusion Matrix")

    # Display the plot
    plt.show()
```

```python
def convert_proba_to_target(y_pred_prob):
    y_pred = np.array(y_pred_prob) >= 0.5
    y_pred = y_pred.astype(int)  # Convert boolean values to integers (0 or 1)

    # Now, y_pred contains binary class labels (0 or 1)
    return y_pred
```
